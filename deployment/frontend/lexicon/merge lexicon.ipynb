```python
import pandas as pd
```

```python
slang = pd.read_csv('slang ke semi baku.csv')
baku = pd.read_csv('slang ke baku.csv')
```

```python
slang = slang[['slang','formal']]
```

```python
slang
```

```python
baku.head()
```

```python
std_word_replace = {}
for i, row in enumerate(baku['slang']):
    std_word_replace[row] = baku['baku'].iloc[i]
```

```python
slang_replace = {}
for i, row in enumerate(slang['slang']):
    slang_replace[row] = slang['formal'].iloc[i]
```

```python
std_word_replace
```

```python
bakus = []
for i, kata in enumerate(slang['slang']):
    try:
        bakus.append(std_word_replace[kata])
    except:
        bakus.append(slang_replace[kata])
```

```python
slang['baku'] = bakus
```

```python
bakus2 = []
for i, kata in enumerate(slang['baku']):
    try:
        bakus2.append(std_word_replace[kata])
    except:
        bakus2.append(kata)
```

```python
slang['baku final'] = bakus2
```

```python
slang.drop_duplicates(subset='slang', keep='first', inplace=True)
slang
```

```python
slang.to_csv('baku.csv', index=False)
```
